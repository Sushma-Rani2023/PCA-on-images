```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
oliv=datasets.fetch_olivetti_faces()
```

```python
fig= plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(oliv.images[i],cmap=plt.cm.bone)
plt.show()
```

```python
x=oliv.data
y=oliv.target
```

```python
pca=PCA()
pca.fit(x)
```

```python
pca.components_.shape
```

```python
pca.explained_variance_
```

```python
k=0
total= sum(pca.explained_variance_)
current_sum= 0
while current_sum/total <0.95:
    current_sum+=pca.explained_variance_[k]
    k=k+1
k
```

```python
pca=PCA(n_components=123,whiten=True)
transformed_data=pca.fit_transform(x)
transformed_data.shape
```

```python
x_approx=pca.inverse_transform(transformed_data)
x_approx.shape
```

```python
x_approx_image=x_approx.reshape(400,64,64)
```

```python
fig= plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(x_approx_image[i],cmap=plt.cm.bone)
plt.show()
```

```python
eigenv=pca.components_
eigenv.shape
```

```python
eigenfaces =eigenv.reshape((123,64,64))
for i in range(64):
    ax =fig.add_subplot(8,8,i+1)
    ax.imshow(eigenfaces[i],cmap=plt.cm.bone)
plt.show()
```
